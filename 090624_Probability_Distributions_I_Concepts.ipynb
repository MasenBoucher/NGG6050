While a little on the nose for this question, an interesting paper by Buzsaki & Mizuseki (2014) examined the fact that many phenomenon in the brain followed a log-normal distribution, and what the relevance of that might be.
https://www.nature.com/articles/nrn3687

Figure 1C was of particular interest.
Figure 1C.png




As described by the authors: "Probability distribution of the magnitude of the pairwise correlation coefficient among pairs of neurons during theta periods in a behavioural task (RUN) and during SWS, recorded at 50 ms time resolution20; only significantly correlated (P < 0.05) cell pairs are shown. Insets show cross-correlograms of poorly synchronized (left) and highly synchronous (right) neuron pairs during SWS."

The data the authors are talking about is log-normally distributed, not just Gaussian distributed. As you can see from Figure 1C, a log-scale is used on the x-axis. For this reason, the scales on my rough replication of these distribution will be a bit off -- I focused on the general shape of the SWS distribution specifically.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
#Code modified from the example from our GitHub and supplements via other resources
#Roughly replicating the RUN distribution from figure above

mu = 0.5
sigma = 0.5
N = 10000

# Get samples
samples = np.random.normal(mu, sigma, N)

# plot histogram for a given number of bins (using trapz to approximate pdf)
Nbins=30
count, nbins, ignored = plt.hist(samples, 30, density=True)
plt.plot(nbins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (nbins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')

# labels, ets
plt.title(f'Rough Replication of the RUN Distribution')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend(['Simulated', 'Theoretical'])
plt.show()
```
